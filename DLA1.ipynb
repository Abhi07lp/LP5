```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error ,r2_score
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
boston = pd.read_csv("/content/boston_house_prices.csv")
```

```python
boston
```

```python
X = boston[['LSTAT','RM','PTRATIO']]
```

```python
Y = boston['PRICE']
```

```python
X_train ,  X_test ,Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state =4)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled =scaler.fit_transform(X_train)
X_test_scaled =scaler.transform(X_test)
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train_scaled , Y_train)
```

```python
Y_pred_lr = lr_model.predict(X_test)
```

```python
mse_lr = mean_squared_error(Y_test ,Y_pred_lr)
mae_lr = mean_absolute_error(Y_test ,Y_pred_lr)
r2_lr =r2_score(Y_test ,Y_pred_lr)
```

```python
print("MSE (Linear Regression):", mse_lr)
print("MAE (Linear Regression):", mae_lr)
print("R2 Score (Linear Regression):", r2_lr)
```

```python
model = Sequential(
    [
        Dense(128 ,activation = 'relu' , input_dim =3),
        Dense(64 ,activation = 'relu'),
        Dense(32,activation ='relu'),
        Dense(16,activation ='relu'),
        Dense(1)

        ]
)
```

```python
model.compile(optimizer = 'adam', loss='mse',metrics = ['mae'])
```

```python
history =model.fit(X_train_scaled , Y_train , epochs =100 , validation_split=0.05 ,verbose =1 )
```

```python
Y_pred_nn = model.predict(X_test_scaled)
```

```python
mse_nn , mae_nn = model.evaluate(X_test_scaled , Y_test)
```

```python
print("MSE (Neural Network):", mse_nn)
print("MAE (Neural Network):", mae_nn)
```

```python
new_data = np.array([[0.1 , 10.0 , 5.0]])
```

```python
new_data_scaled = scaler.transform(new_data)
```

```python
prediction = model.predict(new_data_scaled)
```
